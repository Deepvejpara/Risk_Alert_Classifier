```python
import streamlit as st
import numpy as np
import pickle

# Load trained model
model = pickle.load(open("model.pkl", "rb"))

st.set_page_config(page_title="Risk Prediction App")

st.title("💳 Customer Risk Prediction System")

st.write("Enter customer details to predict risk level")

# ---- INPUT FIELDS (edit based on your dataset) ----
age = st.number_input("Age", min_value=18, max_value=100)
income = st.number_input("Income")
loan_amount = st.number_input("Loan Amount")
credit_score = st.number_input("Credit Score")

# ---- PREDICTION ----
if st.button("Predict Risk"):
    
    input_data = np.array([[age, income, loan_amount, credit_score]])
    
    prediction = model.predict(input_data)
    probability = model.predict_proba(input_data)

    if prediction[0] == 1:
        st.error(f"⚠️ High Risk Customer\nProbability: {probability[0][1]:.2f}")
    else:
        st.success(f"✅ Low Risk Customer\nProbability: {probability[0][0]:.2f}")
```
